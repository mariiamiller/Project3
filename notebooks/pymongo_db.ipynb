```python
from pymongo import MongoClient
```

```python
client = MongoClient(port=27017)
db = client.songs_db
```

```python
import pandas as pd
```

```python
df = pd.read_csv('songs_links.csv',encoding = 'latin1')
df.head()
```

```python
song_list = df.to_dict('records')
```

```python
for song in song_list:
    result = db.songs_links.insert_one(song)
    print(song)
```
