```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras.utils import to_categorical
import pandas as pd
import os
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from numpy.random import seed
seed(1)
import tensorflow
```

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df_meta = pd.read_csv('perfs.csv')
df=df_meta[['songid','perf_key']]
df1 = df_meta['songid'].value_counts().reset_index()
arrang = list(df1['index'])
arrang
```

```python
arrang=arrang[:50]
```

```python
array_perf = []
for arr in arrang:
    new_df = df.loc[df['songid']==arr]
    new_df = new_df.head(200)
    new_list = list(new_df['perf_key'])
    for li in new_list:
            array_perf.append(li)
            print(new_list)
```

```python
df_v = df.loc[df['perf_key'].isin(array_perf)]
#df_v2 = df_v.loc[df['songid'].isin(arrang[:10])]\n",
#df_v.head()\n",
#len(df_v)\n",
perf_test = list(df['perf_key'])
df_v.head()
len(df_v)
df_test=df_v
```

```python
len(df_test)
```

```python
sg_array = []
sg_labels = []
for perf in array_perf:
    try:
        path = 'audio/'+perf+'.m4a'
        y, sr = librosa.load(path)
        y= y[sr*30:sr*90]
        sg0 = librosa.stft(y)
        sg_mag, sg_phase = librosa.magphase(sg0)
        sg1 = librosa.feature.melspectrogram(S=sg_mag, sr=sr)
        sg2 = librosa.amplitude_to_db(sg1, ref=np.min)
        print(sg2.shape)
        df2 = df_test.loc[df_test['perf_key']==perf]
        df2 = df2.reset_index()
        class_song = df2['songid'][0]
        if(sg2.shape==(128, 2584)):
            sg_array.append(sg2)
            sg_labels.append(class_song)     
        print(class_song)
    except:
        print('error'+perf)
```

```python
X = np.array(sg_array)
y = np.array(sg_labels)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, random_state=1, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

scalers = {}
for i in range(X_train.shape[1]):
    scalers[i] = StandardScaler()
    X_train[:, i, :] = scalers[i].fit_transform(X_train[:, i, :]) 

for i in range(X_test.shape[1]):
    X_test[:, i, :] = scalers[i].transform(X_test[:, i, :]) 
```

```python
label_encoder = LabelEncoder()
label_encoder.fit(y_train)
encoded_y_train = label_encoder.transform(y_train)
encoded_y_test = label_encoder.transform(y_test)
```

```python
df_test['songid'].nunique()
```

```python
y_train_categorical = to_categorical(encoded_y_train, num_classes = 50)
y_test_categorical = to_categorical(encoded_y_test,num_classes = 50)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Embedding
model = Sequential()

model.add(Dense(50, input_shape=(128, 2584),activation='relu'))
model.add(Flatten())

model.add(Dense(50, activation='relu'))
model.add(Dense(units=50, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
model.fit(
    X_train,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=1
)
```

```python
model.save('megamodel.h5')
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test, y_test_categorical, verbose=2)
print(
    f"Deep Neural Network - Loss tanh(50,50,100): {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss tanh(50,50,100): {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = model.evaluate(
    X_test, y_test_categorical, verbose=2)
print(
    f"Normal Neural Network - Loss tanh(50,50,1000): {model_loss}, Accuracy: {model_accuracy}")
```
